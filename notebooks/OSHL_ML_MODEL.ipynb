```python
import pandas as pd
import numpy as np

# Load files
df = pd.read_csv("dataset.csv")
severity_df = pd.read_csv("Symptom-severity.csv")

# Create symptom -> weight dictionary
severity_map = dict(
    zip(
        severity_df["Symptom"].str.strip(),
        severity_df["weight"]
    )
)

symptom_cols = [col for col in df.columns if "Symptom" in col]

# Calculate total severity score for each disease case
def calculate_score(row):
    score = 0

    for col in symptom_cols:
        symptom = row[col]

        if pd.notna(symptom):
            symptom = str(symptom).strip()
            score += severity_map.get(symptom, 0)

    return score

df["severity_score"] = df.apply(calculate_score, axis=1)

print(df[["Disease", "severity_score"]].head(20))

print("\n")
print(df["severity_score"].describe())
```

```python
print(df["severity_score"].quantile([0.25, 0.50, 0.75]))
```

```python
def severity_level(score):
    if score <= 18:
        return "Low"
    elif score <= 27:
        return "Moderate"
    elif score <= 42:
        return "Serious"
    else:
        return "Emergency"

df["severity_level"] = df["severity_score"].apply(severity_level)

print(df[["Disease", "severity_score", "severity_level"]].head(20))

print("\nSeverity Distribution:\n")
print(df["severity_level"].value_counts())
```

```python
service_mapping = {
    "Low": "Pharmacy",
    "Moderate": "Telemedicine",
    "Serious": "Clinic",
    "Emergency": "Emergency Care"
}

df["service"] = df["severity_level"].map(service_mapping)

print(df[["severity_level", "service"]].head(20))

print("\nService Distribution:\n")
print(df["service"].value_counts())

print("\nMissing Values:")
print(df["service"].isnull().sum())
```

```python
symptom_cols = [col for col in df.columns if "Symptom" in col]

def combine_symptoms(row):
    symptoms = []

    for col in symptom_cols:
        if pd.notna(row[col]):
            symptoms.append(str(row[col]).strip())

    return " ".join(symptoms)

df["combined_symptoms"] = df.apply(combine_symptoms, axis=1)

print(df["combined_symptoms"].head())

print("\nExample:\n")
print(df["combined_symptoms"].iloc[0])
```

```python
df_unique = df.drop_duplicates(subset=["combined_symptoms"]).copy()

print("Original Dataset:", len(df))
print("Unique Dataset:", len(df_unique))

print("\nSeverity Distribution:")
print(df_unique["severity_level"].value_counts())

print("\nSpecialist Distribution:")
print(df_unique["specialist"].value_counts())
```

```python
from sklearn.model_selection import train_test_split

X = df_unique["combined_symptoms"]

y = df_unique["severity_level"]

print("Total Samples:", len(X))

print("\nSeverity Classes:")
print(y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Samples:", len(X_train))
print("Testing Samples:", len(X_test))

print("\nTraining Distribution:")
print(y_train.value_counts())

print("\nTesting Distribution:")
print(y_test.value_counts())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print("Training Shape:", X_train_tfidf.shape)
print("Testing Shape:", X_test_tfidf.shape)

print("\nVocabulary Size:", len(tfidf.vocabulary_))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

baseline_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

baseline_model.fit(X_train_tfidf, y_train)

y_pred = baseline_model.predict(X_test_tfidf)

print("Accuracy:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# Encode labels
le = LabelEncoder()

y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

# Train XGBoost
xgb_model = XGBClassifier(
    n_estimators=100,
    max_depth=5,
    learning_rate=0.1,
    objective="multi:softmax",
    num_class=4,
    random_state=42
)

xgb_model.fit(X_train_tfidf, y_train_enc)

# Predict
y_pred_enc = xgb_model.predict(X_test_tfidf)

# Convert back to labels
y_pred = le.inverse_transform(y_pred_enc)

print("Accuracy:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import joblib

# Load models
disease_model = joblib.load("disease_model.pkl")
specialist_model = joblib.load("specialist_model.pkl")

severity_model = joblib.load("severity_model.pkl")
severity_tfidf = joblib.load("severity_tfidf.pkl")


def healthcare_recommendation(symptoms_text):

    # Disease Prediction
    disease = disease_model.predict([symptoms_text])[0]

    # Specialist Prediction
    specialist = specialist_model.predict([symptoms_text])[0]

    # Severity Prediction
    symptoms_vector = severity_tfidf.transform([symptoms_text])
    severity = severity_model.predict(symptoms_vector)[0]

    # Service Recommendation
    service_mapping = {
        "Low": "Pharmacy",
        "Moderate": "Telemedicine",
        "Serious": "Clinic",
        "Emergency": "Emergency Care"
    }

    service = service_mapping[severity]

    return {
        "disease": disease,
        "severity": severity,
        "specialist": specialist,
        "service": service
    }
```
